Create dashboards using the data from "/content/tested.csv".

```python
import pandas as pd

df = pd.read_csv('/content/tested.csv')
```

```python
display(df.head())
display(df.info())
display(df.isnull().sum())
```

## Explore the data

### Subtask:
Explore the data by displaying the first few rows, checking the data types, and looking for missing values.


```python
display(df.head())
display(df.info())
display(df.isnull().sum())
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Fare'].fillna(df['Fare'].median(), inplace=True)
df.drop('Cabin', axis=1, inplace=True)
display(df.isnull().sum())
```

```python
display(df.info())
df['Sex'] = df['Sex'].astype('category')
df['Embarked'] = df['Embarked'].astype('category')
display(df.info())
```

## Analyze the data

### Subtask:
Perform some basic analysis to understand the data distribution and relationships between columns.


```python
display(df.describe())
display(df['Sex'].value_counts())
display(df['Embarked'].value_counts())
display(df.corr(numeric_only=True))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the figure size for better readability
plt.figure(figsize=(18, 15))

# 1. Histogram of 'Age'
plt.subplot(3, 3, 1)
sns.histplot(df['Age'].dropna(), kde=True)
plt.title('Distribution of Age')

# 2. Bar plot of 'Survived'
plt.subplot(3, 3, 2)
sns.countplot(x='Survived', data=df)
plt.title('Distribution of Survived')
plt.xticks([0, 1], ['Not Survived', 'Survived'])

# 3. Bar plot of 'Pclass'
plt.subplot(3, 3, 3)
sns.countplot(x='Pclass', data=df)
plt.title('Distribution of Passenger Class')

# 4. Bar plot of 'Sex'
plt.subplot(3, 3, 4)
sns.countplot(x='Sex', data=df)
plt.title('Distribution of Sex')

# 5. Bar plot of 'Embarked'
plt.subplot(3, 3, 5)
sns.countplot(x='Embarked', data=df)
plt.title('Distribution of Embarkation Point')

# 6. Scatter plot of 'Age' vs 'Fare'
plt.subplot(3, 3, 6)
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title('Age vs Fare')

# 7. Box plot of 'Fare' by 'Pclass'
plt.subplot(3, 3, 7)
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare Distribution by Passenger Class')

# 8. Box plot of 'Age' by 'Survived'
plt.subplot(3, 3, 8)
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age Distribution by Survived Status')
plt.xticks([0, 1], ['Not Survived', 'Survived'])

# 9. Box plot of 'Fare' by 'Survived'
plt.subplot(3, 3, 9)
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare Distribution by Survived Status')
plt.xticks([0, 1], ['Not Survived', 'Survived'])

plt.tight_layout()
plt.show()
```

```python
import panel as pn

# Assuming the plots from the previous step are stored in matplotlib figure objects
# and that you want to create a dashboard with them.
# In the previous step, we used subplots within a single figure.
# To use panel effectively, it's better to create individual figures for each plot.

# Re-generating the plots as individual figures for panel
plot_age_hist = sns.histplot(df['Age'].dropna(), kde=True).get_figure()
plt.close(plot_age_hist) # Close the figure to prevent it from displaying twice

plot_survived_bar = sns.countplot(x='Survived', data=df).get_figure()
plt.close(plot_survived_bar)

plot_pclass_bar = sns.countplot(x='Pclass', data=df).get_figure()
plt.close(plot_pclass_bar)

plot_sex_bar = sns.countplot(x='Sex', data=df).get_figure()
plt.close(plot_sex_bar)

plot_embarked_bar = sns.countplot(x='Embarked', data=df).get_figure()
plt.close(plot_embarked_bar)

plot_age_fare_scatter = sns.scatterplot(x='Age', y='Fare', data=df).get_figure()
plt.close(plot_age_fare_scatter)

plot_fare_pclass_box = sns.boxplot(x='Pclass', y='Fare', data=df).get_figure()
plt.close(plot_fare_pclass_box)

plot_age_survived_box = sns.boxplot(x='Survived', y='Age', data=df).get_figure()
plt.close(plot_age_survived_box)

plot_fare_survived_box = sns.boxplot(x='Survived', y='Fare', data=df).get_figure()
plt.close(plot_fare_survived_box)


# Create a list of the generated plots
plots = [
    plot_age_hist,
    plot_survived_bar,
    plot_pclass_bar,
    plot_sex_bar,
    plot_embarked_bar,
    plot_age_fare_scatter,
    plot_fare_pclass_box,
    plot_age_survived_box,
    plot_fare_survived_box,
]

# Arrange the plots in a visually appealing layout using panel
# Example: Two rows of plots
dashboard_layout = pn.Column(
    pn.Row(plots[0], plots[1], plots[2]),
    pn.Row(plots[3], plots[4], plots[5]),
    pn.Row(plots[6], plots[7], plots[8])
)

# To serve the panel layout, you would typically run this in a script or a dedicated cell
# dashboard_layout.servable()
# Or to serve directly in a notebook, you can use:
# dashboard_layout.show()
# However, for the purpose of this task within the provided environment,
# we will just create the layout and report success.
# Running .serve() or .show() would block the execution flow.
```
